```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import ElementClickInterceptedException
from time import sleep

edge_service = webdriver.FirefoxService(executable_path='geckodriver.exe')

# Initialize the WebDriver with the service
driver = webdriver.Firefox(service=edge_service)

Ati = 'https://cdn.ime.co.ir/'

driver.get(Ati)
k = 6

SectionsLinks = [f'#futureView > div > uib-accordion > div > div > div:nth-child({i})' for i in range(1,7)]

Pesteh =        [f'div.panel-open:nth-child(6) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i})' for i in range(2,k)]
PestehTable = [f'div.panel-open:nth-child(6) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i}) > div:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > table:nth-child(1)' for i in range(2,k)]

Kahroba = [f'div.panel-open:nth-child(5) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i})' for i in range(2,k)]
KahrobaTable = [f'div.panel-open:nth-child(5) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i}) > div:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > table:nth-child(1)' for i in range(2,k)]

Silver =  [f'div.panel-open:nth-child(4) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i})' for i in range(2,k)]
SilverTable = [f'div.content > div:nth-child(4) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i}) > div:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > table:nth-child(1)' for i in range(2,k)]

Gold =    [f'div.panel-open:nth-child(3) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i})' for i in range(2,k)]
GoldTable = [f'div.content > div:nth-child(3) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i}) > div:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > table:nth-child(1)' for i in range(2,k)]

Safaron = [f'div.panel-open:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i})' for i in range(2,k)]
SafaronTable = [f'div.content > div:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i}) > div:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > table:nth-child(1)' for i in range(2,k)]

Lotus =   [f'#futureView > div > uib-accordion > div > div > div:nth-child(1) > div > div.panel-collapse.in.collapse > div > div:nth-child({i})' for i in range(2,k)]
LotusTable = [f'div.ng-isolate-scope:nth-child(1) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > div:nth-child({i}) > div:nth-child(2) > div:nth-child(1) > div:nth-child(2) > div:nth-child(1) > table:nth-child(1)' for i in range(2,k)]

Indexes = [Lotus, Safaron, Gold, Silver, Kahroba, Pesteh]
TablesIndex = [LotusTable, SafaronTable, GoldTable, SilverTable, KahrobaTable, PestehTable]

Contract_code = []
Demand_volume, Demand_price = [], []
Supply_volume, Supply_price = [], []
df = pd.DataFrame()
# Find and click on the table elements to reveal the data
for j in range(len(Indexes)):


    for i in range(0,5):

        Result = []

        try:

            #table_elements1 = WebDriverWait(driver, 60).until(EC.visibility_of_all_elements_located((By.CSS_SELECTOR, SectionsLinks[j])))
            #table_elements1[0].click()

            table_elements2 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, Indexes[j][i])))
            #driver.execute_script("arguments[0].click();", WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, Indexes[j][i]))))

            #for table_element in table_elements:
            #    table_element.click()
            #table_elements2[0].click()

            driver.execute_script("scrollBy(0,-30);")
            sleep(2)

            driver.execute_script("scrollBy(0,45);")
            sleep(2)

            driver.execute_script("scrollBy(0,-15);")
            sleep(2)

            table_elements2.click()

            
            table_data = WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, TablesIndex[j][i])))
            for table in table_data:
                #print(table.text)
                Result.append(table.text)

            result_list = Result[0].split('\n')

            Contract_code = Contract_code + [result_list[4].split(' ')[0],result_list[4].split(' ')[0],result_list[4].split(' ')[0]]
            Demand_volume = Demand_volume + [int(result_list[4].split(' ')[1].replace(",", "")), int(result_list[5].split(' ')[0].replace(",", "")), int(result_list[15].split(' ')[0].replace(",", ""))]
            Demand_price = Demand_price + [int(result_list[4].split(' ')[2].replace(",", "")), int(result_list[5].split(' ')[1].replace(",", "")), int(result_list[15].split(' ')[1].replace(",", ""))]
            Supply_volume = Supply_volume + [int(result_list[4].split(' ')[3].replace(",", "")), int(result_list[5].split(' ')[2].replace(",", "")), int(result_list[15].split(' ')[2].replace(",", ""))]
            Supply_price = Supply_price + [int(result_list[4].split(' ')[4].replace(",", "")), int(result_list[5].split(' ')[3].replace(",", "")), int(result_list[15].split(' ')[3].replace(",", ""))]
            print(result_list[4].split(' ')[0])
        
        except:
            break
    

df['کد قرارداد'] = Contract_code
df['حجم تقاضا'] = Demand_volume
df['قیمت تقاضا'] = Demand_price
df['حجم عرضه'] = Supply_volume
df['قیمت عرضه'] = Supply_price
# After you're done, close the WebDriver
driver.quit()
```
